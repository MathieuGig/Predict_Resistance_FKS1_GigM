# Combine S2 Taxonomy Dataframes

```python
import pandas as pd
import numpy as np
```

```python
def combine(df1, df2):
    
    dataframe1 = pd.read_csv(df1, index_col=0)
    dataframe2 = pd.read_csv(df2, index_col=0)
    
    # Add new column
    dataframe1['Paralog'] = 'Fks1'
    dataframe2['Paralog'] = 'Fks2'
    
    dataframe3 = pd.concat([dataframe1, dataframe2])
    
    dataframe3 = dataframe3[['Species', 'Kingdom', 'Phylum', 'Paralog', 'Hotspot1', 'Hotspot2',
     'GapsHotspot1',
     'GapsHotspot2',
     'HS1_in_oPools',
     'HS2_in_oPools',
     'Is_Human_Pathogen']]
    
    return dataframe3

combinedDF = combine('S2_FKS1_Taxonomy.csv', 'S2_FKS2_Taxonomy.csv')

combinedDF
```

```python
combinedDF.to_csv('S2_taxonomy.csv', index=False)
```
